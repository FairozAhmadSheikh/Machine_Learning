```python
import numpy as np
import matplotlib.pyplot as plt
X=np.arange(10)
Y=(X-5)**2+3
print(X,Y)
# Visualise Data 
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
# Gradient Descent
# Init x with any random value 
x=0
lr=0.1
#Plot Orignal Curve
plt.plot(X,Y)
for i in range(50):
    # Calc Gradient dy/dx = 2*(x-5) + 0 for x-5**2 + 3
    grad=2*(x-5)
    # Update x for inc or dec
    x=x-lr*grad
    y=(x-5)**2+3
    plt.scatter(x,y)
plt.show()
```
