```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Generate Data 
def generateDataset(m):
    X=np.random.randn(m)*10 
    noise=np.random.randn(m)
    y= 3 * X + 2 + 5 * noise
    return X,y

X,y=generateDataset(100)
print(X.shape,y.shape)
```

```python
# Visualize data 
def plotData(X,y,title="Data",color="orange"):
    plt.title(title)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.scatter(X,y,c=color)
    plt.show()
```

```python
# Data Normalization
def normaliseData(X):
    X=(X-X.mean())/X.std()
    return X
plotData(X,y)  # Before Normalization
print(X.mean())
print(X.std())
X=normaliseData(X)
plotData(X,y) # After normalization
print(X.mean())
print(X.std())
```

```python
# Train Test Split
def trainTestSplit(X,y,split=0.8):
    m=X.shape[0]  # How many Rows we have is stored in m 
    data=np.zeros((m,2)) #  rows = m and  colms=2
    data[:,0]=X      # Place X in all rows of data but for column 0
    data[:,1]=y     #Place Y in all rows of data but for column 1
    np.random.shuffle(data)  # Shuffle data 
    split=int(m*split)
    # Training Part
    XT=data[:split,0]
    yT=data[:split,1]

    # Testing Part
    Xt=data[split:,0]
    yt=data[split:,1]

    return XT,yT,Xt,yt

# Fucntion call
XT,yT,Xt,yt=trainTestSplit(X,y)
# Length of Training and testing data 
print(XT.shape,yT.shape)
print(Xt.shape,yt.shape)
# Plotting the Train and test data 
plt.title("Train Test Data")
plt.scatter(XT,yT,color="red",label="Test Data")
plt.scatter(Xt,yt,color="green",label="Train Data")
plt.legend()
plt.show()
```
